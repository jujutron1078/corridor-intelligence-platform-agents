```python
grant_writer_docker_url = "http://localhost:8000"

kplc_agent= "https://innovations.kplcinstitute.ac.ke/case-study-matcher"




base_url = kplc_agent
```

```python
from langgraph_sdk import get_client

client = get_client(url=base_url)

# create a thread
thread = await client.threads.create()
thread_id = thread["thread_id"]

threads = await client.threads.search()
for thread in threads:
    print(thread)
```

```python
# List all assistants
assistants = await client.assistants.search()

assistants
```

```python
assistant_name = assistants[0]['name']
assistant_id = assistants[0]['assistant_id']

print(f"Assistant Name: {assistant_name}")
print(f"Assistant ID: {assistant_id}")

```

```python
input = {
    "messages": [{"role": "user", "content": "I want to onboard a project"}],
}


# create a streaming run
async for chunk in client.runs.stream(
    thread_id, assistant_id, input=input, stream_mode="updates"
):
    print(chunk.data)
```
